```python
import pandas as pd
from sklearn.preprocessing import label_binarize
import numpy as np
```

```python
df_user_log = pd.read_csv('https://drive.google.com/uc?id=1DwEvKOYl0hQmd2muwDa13O17smPPgn0D')
df_user_info = pd.read_csv('https://drive.google.com/uc?id=1wM9LY4rAsMLGe6GUZNQd50R8WHiH9xaf')
```

```python


df_user_log.isnull().any()
df_user_info.isnull().any()
user_action = df_user_log.groupby('user_id').size().to_frame().reset_index()
user_action.columns = ['user_id','user_total_action']
user_action.head()

```

```python
user_gender = df_user_info[['user_id','gender']]
user_gender = user_gender.fillna(2)
user_gender.head()
```

```python


user_activity = df_user_log[['user_id','action_type']].copy()
user_activity_info = label_binarize(np.array(user_activity.action_type), classes=[2,3])
user_activity_info = pd.DataFrame(data = user_activity_info, columns=['purchase','add_to_favorite'])
user_activity['user_purchase']=user_activity_info.purchase
user_activity['user_Add_to_favorite']=user_activity_info.add_to_favorite
user_activity_count = user_activity.groupby('user_id').agg('sum').reset_index()
user_activity_count.columns = ['user_id','action_type','user_purchase_count','user_Add_to_favorite_count']
user_activity_count = user_activity_count.drop(columns=['action_type'])
user_activity_count.head()

```

```python
user_activity_count['purchase_&_favorite_count'] = 0
user_activity_count.loc[(user_activity_count['user_purchase_count'] == 1) & (user_activity_count['user_Add_to_favorite_count'] == 1), 'purchase_&_favorite_count'] = 1
user_activity_count['purchase_&_favorite_count'] = user_activity_count['purchase_&_favorite_count'].fillna(0)
user_activity_count.head()
```

```python
user_activity_count = user_activity_count.drop('both_actions', axis=1)
user_activity_count.head()
```

```python
# assuming your DataFrame is named `df`
result = user_activity_count[user_activity_count['purchase_&_favorite_count'] == 1]

# display the filtered DataFrame
print(result)
```
